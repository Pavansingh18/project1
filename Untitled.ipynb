```python
df.dropna(inplace=True)          # Remove rows with missing values
df.isnull().sum()
```

```python
z_scores_feature1 = np.abs(stats.zscore(df['Suscribers']))
outliers_feature1 = np.where(z_scores_feature1 > 3)
print("Outliers in 'feature1':", df.iloc[outliers_feature1])

z_scores_feature2 = np.abs(stats.zscore(df['Visits']))
outliers_feature2 = np.where(z_scores_feature2 > 3)
print("Outliers in 'feature2':", df.iloc[outliers_feature2])
```

```python
df['Suscribers'] = df['Suscribers'].astype(int)
df['Likes'] = df['Likes'].astype(int)
df['Visits'] = df['Visits'].astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of categories
plt.figure(figsize=(20, 10))
sns.countplot(y='Categories', data=df, order=df['Categories'].value_counts().index)
plt.title('Distribution of Categories')
plt.xlabel('Suscribers')
plt.ylabel('Categories')
plt.show()
```

```python
correlation_matrix = df[['Suscribers', 'Likes', 'Comments']].corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot for subscribers vs likes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Suscribers', y='Likes', data=df)
plt.title('Subscribers vs Likes')
plt.xlabel('Subscribers')
plt.ylabel('Likes')
plt.show()

# Scatter plot for subscribers vs comments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Suscribers', y='Comments', data=df)
plt.title('Subscribers vs Comments')
plt.xlabel('Subscribers')
plt.ylabel('Comments')
plt.show()
```

```python
# Calculate audience size for each category within each country
regional_preferences = df.groupby(['Country', 'Categories'])['Suscribers'].sum().reset_index()

# Pivot the table to get a matrix of audience size by country and category
pivot_table = regional_preferences.pivot(index='Country', columns='Categories', values='Suscribers').fillna(0)

# Display the pivot table
print(pivot_table)
```

```python
# Calculate the average values
average_values = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()

# Display the average values
print(average_values)
```

```python

# Convert the average values to a DataFrame for easy plotting
average_df = average_values.reset_index()
average_df.columns = ['Metric', 'Average']

# Plot the average values
plt.figure(figsize=(5, 3))
sns.barplot(x='Metric', y='Average', data=average_df)
plt.title('Average Number of Subscribers, Views, Likes, and Comments')
plt.xlabel('Metric')
plt.ylabel('Average Value')
plt.show()
```

```python
# Box plots to compare metrics across content categories
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(x='Categories', y='Suscribers', data=df)
plt.title('Subscribers by Category')

plt.subplot(2, 2, 2)
sns.boxplot(x='Categories', y='Visits', data=df)
plt.title('Views by Category')

plt.subplot(2, 2, 3)
sns.boxplot(x='Categories', y='Likes', data=df)
plt.title('Likes by Category')

plt.subplot(2, 2, 4)
sns.boxplot(x='Categories', y='Comments', data=df)
plt.title('Comments by Category')

plt.tight_layout()
plt.show()
```

```python
# Calculate streamer counts by category
category_counts = df['Categories'].value_counts()

# Display the counts
print(category_counts)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Categories', data=df, order=category_counts.index)
plt.title('Distribution of Content Categories')
plt.xlabel('Number of Streamers')
plt.ylabel('Category')
plt.show()
```

```python
avg_metrics_by_category = df.groupby('Categories').agg({
    'Suscribers': 'mean',
    'Visits': 'mean',
    'Likes': 'mean',
    'Comments': 'mean'
}).reset_index()

# Display the average metrics
print(avg_metrics_by_category)
```

```python
performance_metrics = ['Suscribers', 'Visits', 'Likes', 'Comments']
df['Hypothetical_Brand_Collaborations'] = (df['Suscribers'] + df['Visits'] + df['Likes'] + df['Comments']) / 1000
plt.figure(figsize=(14, 10))

for i, metric in enumerate(performance_metrics, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=metric, y='Hypothetical_Brand_Collaborations', data=df)
    plt.title(f'{metric} vs Hypothetical Brand Collaborations')

plt.tight_layout()
plt.show()
```

```python
avg_subscribers = df['Suscribers'].mean()
avg_visits = df['Visits'].mean()
avg_likes = df['Likes'].mean()
avg_comments = df['Comments'].mean()

print(f'Average Subscribers: {avg_subscribers}')
print(f'Average Visits: {avg_visits}')
print(f'Average Likes: {avg_likes}')
print(f'Average Comments: {avg_comments}')
```

```python
above_avg_streamers = df[
    (df['Suscribers'] > avg_subscribers) &
    (df['Visits'] > avg_visits) &
    (df['Likes'] > avg_likes) &
    (df['Comments'] > avg_comments)
]
```

```python
print(above_avg_streamers[['Username', 'Suscribers', 'Visits', 'Likes', 'Comments']])
```

```python
df['Performance_Score'] = df[performance_metrics].sum(axis=1)
```

```python
top_performers = df.sort_values(by='Performance_Score', ascending=False).head(10)
```
